```python
# Extract only sales in year 2022, 2021 sales will come in later

sales_22 = rolling_sales[rolling_sales['SALE DATE'].dt.year == 2022]
sales_22['BOROUGH'].count()
```

```python
#Easement column is empty

#the types need to be changed to categorical for alot fo them 

sales_22.info()
```

```python
#Change type to category for columns 

obj_cats = ['BUILDING CLASS CATEGORY','TAX CLASS AT PRESENT','TAX CLASS AT TIME OF SALE','BOROUGH',
            'BUILDING CLASS AT TIME OF SALE','BUILDING CLASS AT PRESENT','BLOCK','LOT','ZIP CODE','YEAR BUILT']

for colname in obj_cats:
    sales_22[colname] = sales_22[colname].astype('category') 


sales_22.info()
```

```python
#A $0 sale indicates that there was a transfer of ownership without a cash consideration.

zero_22_sales = sales_22[sales_22['sale_price'] == 0]
zero_22_sales['sale_price'].count()
```

```python
# There are sales under a thousand that are suspeciously low

low_22_sales = sales_22[sales_22['sale_price'].between(1,999)]

## difference between $0-999 and just $0

low_22_sales['borough'].count(), zero_22_sales['sale_price'].count()
```

```python
# We can be fairly certain that 0 and under a thousand sales prices do not represent
# the actual value of the building.

# We are going to separate the 0 and low sales prices

sales_22 = sales_22[~sales_22['sale_price'].between(0,999)]
sales_22['sale_price'].count()
```

```python
# Let's analyze our other quantitative columns and their relationship to price

#take out the zero's and big outliers so we can visualize better

non_zero_features = sales_22[(sales_22['sale_price']<49999999) & (sales_22['residential_units'].between(1,200))
                          & (sales_22['commercial_units'].between(1,25)) & (sales_22['total_units'].between(1,200))
                          & (sales_22['land_square_feet'].between(1,50000))
                          & (sales_22['gross_square_feet'].between(1,10000))]
non_zero_features.count()
```

```python
# define a function for scatterplots


def scatterplots(columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x = col, y = 'sale_price', data=non_zero_features, alpha=0.5)
        ax.set(xlabel=col, ylabel='Sale price')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
# relationship between sales price and quantitative columns 

#gross sq feet is most defined relationship


features = ['residential_units','commercial_units', 'total_units', 'land_square_feet',
       'gross_square_feet']

scatterplots(features,figsize=(24,5))
```

```python
vacancy.rename(columns={'vacant_6/30_or_date_sold_if_earlier':'vacant_6/30'},inplace=True)
```

```python
bor_vacancy = vacancy.groupby(by=['borough','reporting_year']).agg({'vacant_on_12/31':'sum',
                                                                    'vacant_6/30':'sum'})
bor_vacancy.reset_index(inplace=True)
```

```python
bor_totals = vacancy.groupby(by=['borough','reporting_year']).agg({'vacant_on_12/31':'count'})
bor_totals.reset_index(inplace=True)
```

```python
bor_vac_rate = pd.concat([bor_totals['vacant_on_12/31'],bor_vacancy],axis=1)
bor_vac_rate.columns = ['total','borough','reporting_year','vacant_on_12/31','vacant_6/30']
bor_vac_rate
```

```python
bor_vac_rate['vacant_12/31_perc'] = bor_vac_rate['vacant_on_12/31']/bor_vac_rate['total']*100
bor_vac_rate['vacant_6/31_perc'] = bor_vac_rate['vacant_6/30']/bor_vac_rate['total']*100
bor_vac_rate.drop(columns=['vacant_on_12/31','vacant_6/30','total'],inplace=True)
bor_vac_rate
```
