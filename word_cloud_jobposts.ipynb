```python
## word cloud project

## I am going to put 10-15 job descriptions that I am interested in in a document 

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd

```

```python
df = pd.read_csv("job_post_desc.csv")
df.head()
```

```python
df_melt = pd.melt(df)
df_melt.dropna(inplace=True)
df_melt.tail()
```

```python
df_melt['value'] = df_melt['value'].str.lower()
```

```python
df_melt['value'].value_counts()
```

```python
## add additional words as deemed necessary 

## I saw the wrods I am adding now in the word cloud, so I came back to take them out
## they are too general in my opinion 
extra_sw = ['using','analyst','position','duties','new','work','will','staff','responsibilities','including',
            'working','support','provide','tasks','experience','skills']
```

```python
stopwords = list(STOPWORDS) + extra_sw
stopwords[-6:-1]
```

```python
#make into a text file 

text = " ".join(i for i in df_melt['value'])
```

```python
#from column to a text 

text[0:300]
```

```python
#creating graph


wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                max_words=22, 
                min_font_size = 10).generate(text)
 
# plot the WordCloud image                      
plt.figure(figsize = (10, 10), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
